```python
fig = plt.figure()
ax = fig.gca(projection='3d')
x = np.linspace(-2.5,2.5,num=2**16)
ax.plot(x, np.zeros(len(x)), x+2, c='blue')
ax.plot(x, np.zeros(len(x)), 2*x, c='purple')
ax.plot(x, np.zeros(len(x)), x**2, c='g')
ax.plot(x, np.zeros(len(x)), x**x, c='r')
xn = np.linspace(-2.5,0,num=2**16)
#for n in range(1,55):
n=1
ax.plot(xn, abs(xn)**xn * np.sin(n*xn*np.pi), abs(xn)**xn * np.cos(n*xn*np.pi),c='r')
ax.plot(xn, np.zeros(len(xn)), abs(xn)**xn * np.cos(n*xn*np.pi),c='grey')
ax.plot(xn, abs(xn)**xn * np.sin(n*xn*np.pi), np.zeros(len(xn)),c='grey')
ax.scatter(2,0,4, c='gold')

ax.set_xlim(-2.5,2.5)
ax.set_ylim(-2.5,2.5)
ax.set_zlim(-2.5,2.5)
plt.show()
```

```python
def Re3(x):
    n = 1
    term1 = np.exp(-abs(x)**x * n * np.pi * np.sin(n*np.pi*x))
    term2 = abs(x)**(abs(x)**x * np.cos(n*np.pi*x))
    term3 = np.cos(abs(x)**x * n * np.pi * np.cos(n*np.pi*x) + (abs(x)**x * np.log(x**2) * np.sin(n*np.pi*x))/2)
    return term1*term2*term3
def Im3(x):
    n = 1
    term1 = np.exp(-abs(x)**x * n * np.pi * np.sin(n*np.pi*x))
    term2 = abs(x)**(abs(x)**x * np.cos(n*np.pi*x))
    term3 = np.sin(abs(x)**x * n * np.pi * np.cos(n*np.pi*x) + (abs(x)**x * np.log(x**2) * np.sin(n*np.pi*x))/2)
    return term1*term2*term3
fig = plt.figure()
ax = fig.gca(projection='3d')
xp = np.linspace(0,2.2,num=2**16)
ax.plot(xp, np.zeros(len(xp)), xp**xp**xp, c='r')
xn = np.linspace(-2.5,-0,num=2**16)
ax.plot(xn, Im3(xn), Re3(xn), c='purple')
ax.plot(xn, Im3(xn), np.zeros(len(xn)), c='gray')
ax.plot(xn, np.ones(len(xn)), Re3(xn), c='gray')
#ax.set_xlim(-2.5,2.5)
#ax.set_ylim(-3,2.5)
#ax.set_zlim(-2.5,2.5)
plt.show()
```
