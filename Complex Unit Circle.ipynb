```python
import numpy as np
from numpy import real, imag, pi, e
import matplotlib.pyplot as plt
from scipy.optimize import fmin, fsolve
from scipy.special import lambertw as W
#from mpl_toolkits.mplot3d import Axes3D
```

```python
def cis(z): return np.exp(1j*z)
def cis2(z): return cis(z)**cis(z)
def cis3(z): return cis(z)**cis(z)**cis(z)
def cis3(z): return cis(z)**cis(z)**cis(z)
def cis4(z): return cis(z)**cis(z)**cis(z)**cis(z)
def cis5(z): return cis(z)**cis(z)**cis(z)**cis(z)**cis(z)
def inf(z): return -W(-np.log(z)) / np.log(z)
def ssqrt(z): return np.real(np.log(z)/W(np.log(z)))
def cisinf(z): return inf(cis(z))
def recis2(z): return np.real(cis2(z))
def imcis2(z): return np.imag(cis2(z))
```

```python
extr1 = fmin(lambda x: -imcis2(x), x0=pi/2, xtol=1e-16, ftol=1e-16)[0]
extr2 = fmin(lambda x: -imcis2(x), x0=3*pi/2, xtol=1e-16, ftol=1e-16)[0]
print(imcis2(extr1))
print(imcis2(extr2))
```
