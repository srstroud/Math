```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
phi = (1+np.sqrt(5))/2
def F(n):
    return ((1+np.sqrt(5))**n-(1-np.sqrt(5))**n)/(2**n*np.sqrt(5))
def lowF(n):
    return np.floor(np.log(np.sqrt(5)*n)/np.log(phi))
def upF(n):
    return np.ceil(np.log(np.sqrt(5)*n)/np.log(phi))
```

```python
lowF(4**256)
```

```python
x=np.arange(0,7)
plt.plot(x,F(x),'bo')
plt.show()
```

```python
x=np.arange(0,7)
lowratio = F(lowF(x**x))/x**x
upratio = F(upF(x**x))/x**x
plt.plot(x, lowratio, 'bo')
plt.plot(x, upratio, 'ro')
plt.show()
```
