```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lambertw
import math
```

```python
def ssqrt(x):
    if x >=1:
        ssqrt = math.log(x) / lambertw(math.log(x))
        return np.real(ssqrt)
    else:
        print('Input must be greater than 1.')
        
def inf(x):
    ssqrt = np.log(x) / lambertw(np.log(x))
    return 1/ssqrt(1/x)
```

```python
class Tet(object):
    """Don't forget to write docstrings!"""
    def __init__(self, arg1):
        self.arg1 = arg1
        
    def ssqrt(x):
        if x >=1:
            ssqrt = np.exp(lambertw(np.log(x)))
            #ssqrt = np.log(x) / lambertw(np.log(x))
            return np.real(ssqrt)
        else:
            print('Input must be greater than 1.')
        
    def inf(x):
        inf = -lambertw(-np.log(x)) / np.log(x)
        if 1/np.e <= x <= np.e:
            return np.real(inf)
        else:
            print('Input must be between 1/e and e^1/e.')
```

```python
import matplotlib.pyplot as plt
eta = (np.e)**(1/np.e)
x = np.linspace(1/np.e, eta, num=1000)
infs = []
for i in range(0, len(x)):
    inf = Tet.inf(x[i])
    infs.append(inf)
    
plt.plot(x, infs, c='r')
plt.plot(1/np.e , Tet.inf(1/np.e),'ro')
plt.plot(eta, Tet.inf(eta),'ro')
plt.grid()
#plt.xlim(0.25,1.5)
#plt.ylim(0.5,3)
plt.show()
```

```python
def tet(x):
    upper_lim = ssqrt(2**65536)
    if 0<= x <= upper_lim:
        return x**x
    if x > upper_lim:
        print('Result too large! Returning number of decimal digits:') 
        return round(math.log10(x**x))
    if round(math.log10(x)) > upper_lim:
        print('Result WAY too large! Returning number of decimal digits IN number of decimal digits:')
        return round(math.log10( round(math.log10(x**x))))
    if x == 1j:
        return np.real(1j**1j)
```

```python
def name(base,height):
    suffix = ['maz', 'wan', 'naeg', 'jinn', 'miu', 'torr', 'chaas', 'korsh', 'baij']
    prefix = [0, '', 'tetra', 'tetre', 'tetri', 'tetro', 'tetru', 'tetry', 'tetraca', 'tetrebe'] 
    if base == 1:
        return suffix[1]
    if height == 1:
        return suffix[base]
    if base == height:
        return 'penta'+suffix[base]
    else:
        print(str(base)+'[IV]'+str(height)+ ' = ' + str(prefix[height] + suffix[base]))
        #return prefix[height] + suffix[base]
```

```python
def dodec(n):
    return int(n*((261*n**3)-(504*n**2)+(283*n)-38)/2)
```
